```python
from PIL import Image
```

```python
def histogram_analysis(path):
    with Image.open(path) as img:
        histogram = img.histogram()
    return histogram
```

```python
origin_path = str(input("Enter origin image path: "))
encrypted_path = str(input("Enter encrypted image path: "))
ori_histogram = histogram_analysis(origin_path)
print(ori_histogram.__len__())
enc_histogram = histogram_analysis(encrypted_path)
import matplotlib.pyplot as plt

# Separate the histograms of the color channels
ori_histogram_r = ori_histogram[:256]
ori_histogram_g = ori_histogram[256:512]
ori_histogram_b = ori_histogram[512:]

enc_histogram_r = enc_histogram[:256]
enc_histogram_g = enc_histogram[256:512]
enc_histogram_b = enc_histogram[512:]

# Plot the histograms
plt.figure(figsize=(16, 8), clear=True)
plt.title('Original Image Histogram')
plt.subplot(2, 3, 1)
plt.bar(range(256), ori_histogram_r, color='red', alpha=0.7)
plt.subplot(2, 3, 2)
plt.bar(range(256), ori_histogram_g, color='green', alpha=0.7)
plt.subplot(2, 3, 3)
plt.bar(range(256), ori_histogram_b, color='blue', alpha=0.7)


plt.subplot(2, 3, 4)
plt.bar(range(256), enc_histogram_r, color='red', alpha=0.7)
plt.subplot(2, 3, 5)
plt.bar(range(256), enc_histogram_g, color='green', alpha=0.7)
plt.title('Encrypted Image Histogram')
plt.subplot(2, 3, 6)
plt.bar(range(256), enc_histogram_b, color='blue', alpha=0.7)


plt.show()
```
